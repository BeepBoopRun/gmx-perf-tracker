```python
import matplotlib.pyplot as plt
import matplotlib.dates as dt
from dateutil.parser import parse
```

```python
inputfile = open('timestamped_step6.0_minimization.log')

dates = []
steps = []

lines = ["", ""]
#read when 3 last lines are
#date
#steps
#number

for line in inputfile:
    lines.append(line)
    if ":" in lines[0] and "Step" in lines[1] and "." in lines[2]:
        dates.append(parse(lines[0]))
        steps.append(int(str(lines[2]).split()[0]))
    lines = lines[1:]

inputfile = open('timestamped_step6.0_minimization2.log')

dates2 = []
steps2 = []

lines = ["", ""]

for line in inputfile:
    lines.append(line)
    if ":" in lines[0] and "Step" in lines[1] and "." in lines[2]:
        dates2.append(parse(lines[0]))
        steps2.append(int(str(lines[2]).split()[0]))
    lines = lines[1:]


dates2 = [date - dates2[0] + dates[0] for date in dates2]
print(len(dates2))
print(len(dates))


plt.plot(dates, steps)
plt.plot(dates2, steps2)
```
