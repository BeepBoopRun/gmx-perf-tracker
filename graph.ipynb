```python
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
#location of file generated by the script
inputfile = open('timestamped_step6.0_minimization3.log')

dates = []
steps = []

lines = ["", ""]
#correct when 3 last lines are
#date
#steps
#number

for line in inputfile:
    lines.append(line)
    if ":" in lines[0] and "Step" in lines[1] and "." in lines[2]:
        dates.append(parse(lines[0]))
        steps.append(int(str(lines[2]).split()[0]))
    lines = lines[1:]

print("Steps registered: " + str(len(dates)))
print("Steps in simulation: " + str(steps[-1]))
plt.figure().set_figwidth(10)
plt.plot(dates, steps)
plt.show()
```
